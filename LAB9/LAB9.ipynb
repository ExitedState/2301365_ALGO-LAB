```python
def KMP_search(pattern, text, direction='LR'):
    """
    Perform a Knuth-Morris-Pratt (KMP) search to find all occurrences of a pattern in a text.

    Parameters:
    - pattern (str): The pattern to search for.
    - text (str): The text to search within.
    - direction (str): Search direction, 'LR' (left-to-right) or 'RL' (right-to-left).

    Returns:
    - list of tuples: Each tuple contains the index of the first character of an occurrence and the search direction.
    """
    if direction == 'RL':
        # Reverse only the pattern for right-to-left search
        pattern = pattern[::-1]

    # Create the longest prefix suffix (LPS) array
    lps = [0] * len(pattern)
    length = 0  # length of the previous longest prefix suffix
    i = 1

    # Preprocess the pattern to create the LPS array
    while i < len(pattern):
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1

    # Start searching for the pattern in the text
    i = 0  # index for text
    j = 0  # index for pattern
    results = []

    while i < len(text):
        if pattern[j] == text[i]:
            i += 1
            j += 1
        if j == len(pattern):
            # Found a match, append the index and direction to results
            if direction == 'RL':
                # Calculate the index for the original text direction
                results.append((i, direction))
            else:
                results.append((i-j+1, direction))
            j = lps[j - 1]
        elif i < len(text) and pattern[j] != text[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1

    return results, lps
```

```python
# Perform KMP search in both directions
results_lr,lps_lr= KMP_search(pattern, text, 'LR')
results_rl,lps_rl = KMP_search(pattern, text, 'RL')

# Combine results and sort them based on the index
combined_results = results_lr + results_rl
print(*lps_lr)
# print(*lps_rl)

print(len(combined_results))
# Print the combined results
for result in combined_results:
    print(f"{result[0]} {result[1]}")
```
